```python
import random
import numpy as np
import plotly.graph_objects as go
```

```python
#help de random doet kut

N = 100 #people
start_cash = 100
rounds = 100
people = [[start_cash] for i in range(N)]
people_alt = [[start_cash] for i in range(N)]

def func_inv_round(input_people, input_people_alt): #a single round of investments
    
    for idx in range(N):
        q = random.random()
        p = 0.2
        if q < 0.5:      
            p1 = 1+p #random.random() #personal profit is based on how many % of their wealth is invested #maybe kijken of normal dist beter werkt hier?
            p2 = 1+p
        else:
            p1 = 1-p
            p2 = 1/(1+p)
        
        input_people[idx].append(input_people[idx][-1]*p1) #add new entry to the persons money record
        
        input_people_alt[idx].append(input_people_alt[idx][-1]*p2) #add new entry to the persons money record
        
    return input_people, input_people_alt #return the list of people with new balances

def func_investments(r): #runs r rounds of investments and plots them
    global people #altering the original variable
    global people_alt
    
    for i in range(r):
        people, people_alt = func_inv_round(people, people_alt)
    
    fig = go.Figure()
    for i in range(N):
        fig.add_trace(go.Scatter(x = range(r), y = people[i]))
    fig.show() 
    
    fig_alt = go.Figure()
    for i in range(N):
        fig_alt.add_trace(go.Scatter(x = range(r), y = people_alt[i]))
    fig_alt.show()  
    
func_investments(rounds)
```

```python
def func_powerlaw_t(t, ret=True, plot=False, log=False):
    money_at_t = sorted([people[i][t] for i in range(N)])[::-1]
    return money_at_t
    
def func_powerlaw_t_alt(t, ret=True):
    money_at_t = sorted([people_alt[i][t] for i in range(N)])[::-1]
    return money_at_t

def list_average(lst):
    t_avg = sum(lst) / len(lst)
    return t_avg

def list_median(lst):
    h = int(len(lst)/2)
    return lst[h]
    
def market_avg(set):
    market = []
    market_alt = []
    for i in range(rounds):
        market.append(list_average(func_powerlaw_t(i)))
        market_alt.append(list_average(func_powerlaw_t_alt(i)))
    
    mkt_plot = go.Figure()
    mkt_plot.add_trace(go.Scatter(x = range(rounds), y = market))
    mkt_plot.add_trace(go.Scatter(x = range(rounds), y = market_alt))
    mkt_plot.show()
    
def market_mdn(set):
    market = []
    market_alt = []
    for i in range(rounds):
        market.append(list_median(func_powerlaw_t(i)))
        market_alt.append(list_median(func_powerlaw_t_alt(i)))
    
    mkt_plot = go.Figure()
    mkt_plot.add_trace(go.Scatter(x = range(rounds), y = market))
    mkt_plot.add_trace(go.Scatter(x = range(rounds), y = market_alt))
    mkt_plot.show()
    
market_avg(people)
market_mdn(people)
```
